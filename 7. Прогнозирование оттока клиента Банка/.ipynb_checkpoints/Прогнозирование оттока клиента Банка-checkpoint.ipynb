## Исследование задачи

### Преобразуем данные и разобъем на выборки

```python
#для построения модели ML посмотрим на целевой признак, в частности на их распределение/ соотношение
sns.countplot(x=data['exited'], hue=data['exited'])
plt.show()
print(data.groupby('exited').size())
print()
print('ушли из банка -', '{}%'.format((data.query('exited == 1')['exited'].count() / data.shape[0]).round(2)))
print('не ушли из банка -', '{}%'.format((data.query('exited == 0')['exited'].count() / data.shape[0]).round(2)))
```

Видим, что есть дисбаланс классов в пропорции 20% к 80%, то есть 1:4.

```python
#в данных присутствуют категориальные признаки, следует преобразовать их в численные
#применим технику прямого кодирования OHE (One-Hot Encoding)
data_ohe = pd.get_dummies(data, drop_first=True)
```

```python
#разобъем данные на выборки тренировачная - валидационная - тестовая : 60 - 20 - 20
#определим признаки (фичи) - X и целевой признак - y
X = data_ohe.drop(columns=['exited'])
y = data_ohe['exited']

#разобьем данные на обучающую, валидационную и тестовую
#выделим под тест и валидацию 40% данных
train_X, test_val_X, train_y, test_val_y = train_test_split(X, y, test_size=0.4, random_state=True)

#выделим валидациооную и тестовую части
#разделим test на валидацию и тестовую выборки
val_X, test_X, val_y, test_y = train_test_split(test_val_X, test_val_y, test_size=0.5, random_state=True)
```

```python
#убедимся в разделении данных на выборки
print(train_X.shape)
print(test_X.shape)
print(val_X.shape)
```

```python
#произведем стандартизацию численных данных, которые ранее определили как переменную num_feat
scaler = StandardScaler()
scaler.fit(train_X[num_feat])
train_X[num_feat] = scaler.transform(train_X[num_feat])
val_X[num_feat] = scaler.transform(val_X[num_feat])
test_X[num_feat] = scaler.transform(test_X[num_feat])
```

### Рассмотрим модель дерево решений с классовым дисбалансом

```python
%%time

f1_model_dtc = []
best_model_dtc = None
best_result = 0
best_depth = 0
for depth in range(1, 50):
    model_dtc = DecisionTreeClassifier(random_state=12345\
                                       , max_depth=depth)
    model_dtc.fit(train_X, train_y)
    predictions = model_dtc.predict(val_X)
    result = f1_score(val_y, predictions)
    f1_model_dtc.append(result.round(2))
    if result > best_result:
        best_model_dtc = model_dtc
        best_result = result
        best_depth = depth

probabilities_dtc = best_model_dtc.predict_proba(val_X)
probabilities_one_valid_dtc = probabilities_dtc[:, 1]
auc_roc_dtc = roc_auc_score(val_y, probabilities_one_valid_dtc)
print('AUC-ROC модели',auc_roc_dtc)
print('F1 score лучшей модели на валидационной выборке:', best_result.round(2))
print('Глубина дерева у лучшей модели:', best_depth)
print(f1_model_dtc)
```

### Рассмотрим модель случайного леса с классовым дисбалансом

```python
%%time

f1_model_rfc = []
best_model_rfc = None
best_result = 0
best_depth = 0
for est in range(10, 200, 10):
    f1_model_rfc.append('|')
    for depth in range(1, 20):
        model_rfc = RandomForestClassifier(random_state=23456\
                                           , n_estimators=est\
                                           , max_depth=depth)
        model_rfc.fit(train_X, train_y)
        predictions = model_rfc.predict(val_X)
        result = f1_score(val_y, predictions)
        f1_model_rfc.append(result.round(2))
        if result > best_result:
            best_model_rfc = model_rfc
            best_result = result
            best_est = est
            best_depth = depth

probabilities_rfc = best_model_rfc.predict_proba(val_X)
probabilities_one_valid_rfc = probabilities_rfc[:, 1]
auc_roc_rfc = roc_auc_score(val_y, probabilities_one_valid_rfc)
print('AUC-ROC модели',auc_roc_rfc)
print('F1 score лучшей модели на валидационной выборке:', best_result.round(2))
print('Количество деревьев:', best_est, 'Глубина дерева у лучшей модели:', best_depth)
print(f1_model_rfc)
```

### Произведем балансировку классов, путем увеличения выборки

```python
#функция увеличения выборки
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=True)
    
    return features_upsampled, target_upsampled
```

```python
X_upsampled, y_upsampled = upsample(train_X, train_y, 4)
```

### Произведем балансировку классов, путем уменьшения выборки

```python
#функция уменьшения выборки
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
X_downsampled, y_downsampled = downsample(train_X, train_y, 0.25)
```
